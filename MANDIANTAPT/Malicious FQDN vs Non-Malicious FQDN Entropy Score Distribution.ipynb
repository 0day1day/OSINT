#Mandiant APT1 Domain Names & Entropy Score Distribution 

Is there any relaitonship between entropy scores of known malicious fully qualified domain names? The data is inconclusive at this stage. Additionally, with the Mandiant APT1 data set is small; however it is a know attributed Computer Network Exploit network utilized by APT1 actors. Perhaps entropy score interpretation is useless at discovering unknown malicious exploit or C2 FQDN's. Requires more research. Nevertheless, the data is made available to whomever is interested with this spit ball analysis. 

```python
import requests 
import math
from uuid import uuid4
from collections import Counter
from pandas import DataFrame


def getData(url_string):
    """Get CSV Mandiant Data Set - output list data structure"""
    response = requests.get(url_string)
    iterResponse = response.iter_lines()
    next(iterResponse)
    for line in iterResponse:
        yield line.split(',')

 
def entropy(string_name):
    p, lns = Counter(string_name), float(len(string_name))
    return -sum( count/lns * math.log(count/lns, 2) for count in p.values())

        
fqdn_list = []
url = "https://raw.githubusercontent.com/alienone/OSINT/master/MANDIANTAPT/APT-Maxmind-Enrichment-Product-2013-07-14-09-25-42.csv"

for items in getData(url):
        if len(items[0]) > 0:
            fqdn_list.append(items[0])

unique_list = list(set(fqdn_list))
product_list = []
for item in unique_list:
    product_list.append({'FQDN': item, 'Entropy': entropy(item)})
print "Number of FQDN's " + str(len(product_list))
dfm = DataFrame(product_list)
```

```python
dfm = dfm.sort('Entropy', ascending=False)
dfm.index = range(1, len(dfm) + 1)
dfm.head(10)
```

```python
dfm.tail(10)
```

```python
dfm.describe()
```

```python
dfm.plot(figsize = (20,9), title=("Mandiant APT1 Domain Names & Entropy Score Distribution"))
```
